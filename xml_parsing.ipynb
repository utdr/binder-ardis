```python
def isPart(line):
    if len(line)==0:
        return False
    else:
        sline = line.split(r".")
        return re.match("[A-Z]", sline[0])

isPart("1. Les formules courantes")
```

```python
x = "$sdfqsf\tmlj*lkj"

# échappe la liste de caractères (ajout de l'antislash)
for caractère in ["$", "^", ".", "*", "+", "-", "?"]:
    x = x.replace(caractère, f"\{caractère}")
print(x)
```
